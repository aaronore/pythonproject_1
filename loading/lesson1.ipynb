```python
import requests
from io import BytesIO
from PIL import Image
imageUrl = 'https://cdn.britannica.com/71/100371-050-A69E9BD1/Brown-bear.jpg'

try:
    response = requests.get(imageUrl,headers=header,stream=True,cookies=cookie)
    response.raise_for_status()
except requests.ConnectionError as e:
    print("網路連線有問題")
    print(e)
except requests.HTTPError as e:
    print("網址和伺服器回傳有問題")
    print(e)

with open('bear.jpg','wb') as file:
    for chunk in response.iter_content(chunk_size=128):
        file.write(chunk)
```
